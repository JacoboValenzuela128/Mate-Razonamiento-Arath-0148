* Edad de Pedro: $2a$ (porque es el doble que la de Ana).
2. Traducir a lenguaje algebraico: "La edad de Pedro ($2a$) más la de Ana ($a$) suman ($+$) 36".
### Ecuación: $$2a + a = 36$$
Resolución (Despeje):
* Primero sumamos las variables similares: $3a = 36$
* Para dejar a la $a$ sola, el 3 que está multiplicando pasa al otro lado dividiendo.
* $a = \frac{36}{3}$$
* a = 12$ (Edad de Ana).
* Para Pedro: $2 \times 12 = 24$.
